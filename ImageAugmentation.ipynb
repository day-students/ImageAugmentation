```
import tensorflow as tf
from tensorflow.keras import layers

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import math

print(tf.version.VERSION)
print(tf.keras.__version__)
```

```
image_path = 'cherries.jpg'

image_string = tf.read_file(image_path)
 
image = tf.image.decode_jpeg(image_string,channels=3)

IMAGE_SIZE = 224

image = tf.image.resize_images(image, (IMAGE_SIZE, IMAGE_SIZE), 
                                    tf.image.ResizeMethod.NEAREST_NEIGHBOR)

with tf.Session() as sess:
      sess.run(tf.global_variables_initializer())
    
      plot = sess.run(image)
      plt.imshow(plot)
```

```
central_image = tf.image.central_crop(image, central_fraction = 0.7)

with tf.Session() as sess:
      sess.run(tf.global_variables_initializer())
    
      plot = sess.run(central_image)
      plt.imshow(plot)
```

```
flip_image = tf.image.flip_left_right(image)

with tf.Session() as sess:
      sess.run(tf.global_variables_initializer())
    
      plot = sess.run(flip_image)
      plt.imshow(plot)
```

```
flip_image = tf.image.flip_up_down(image)

with tf.Session() as sess:
      sess.run(tf.global_variables_initializer())
    
      plot = sess.run(flip_image)
      plt.imshow(plot)
```

```
rotate_image = tf.contrib.image.rotate(image, math.radians(180))

with tf.Session() as sess:
      sess.run(tf.global_variables_initializer())
    
      plot = sess.run(rotate_image)
      plt.imshow(plot)
```

```
move_image = tf.contrib.image.translate(image, translations=[10,10])

with tf.Session() as sess:
      sess.run(tf.global_variables_initializer())
    
      plot = sess.run(move_image)
      plt.imshow(plot)
```

```
bright_image = tf.image.adjust_brightness(image, -0.2)

with tf.Session() as sess:
      sess.run(tf.global_variables_initializer())
    
      plot = sess.run(bright_image)
      plt.imshow(plot)
```
